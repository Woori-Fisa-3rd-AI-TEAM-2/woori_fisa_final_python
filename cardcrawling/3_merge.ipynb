```python
import pandas as pd
credit_file_path = 'creditcard.csv'
check_file_path = 'checkcard.csv'

credit = pd.read_csv(credit_file_path, header=0)
check = pd.read_csv(check_file_path, header=0)

credit['card_type'] = 'credit'
check['card_type'] = 'check'

result = pd.concat([credit, check], ignore_index=True)
```

```python
result.columns
```

```python
result = result[['idx', 'id', 'name', 'brand', 'card_type', 'img_url', 'card_url', 'convenience',
       'cafe', 'restaurant', 'oil', 'movie', 'mart', 'shopping', 'hospital',
       'edu', 'tel', 'car', 'travel', 'transportation']]
```

```python
result.to_csv('card.csv', index = False)
```
